```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import calendar
```

#### **Task 2**: Is average/median monthly delay different during the year? If yes, which are months with the biggest delays and what could be the reason?

Reasons for possible aircraft delays:
   * Weather, natural disasters
   * Congestion
   * Human errors, malfunctions, technical issues
   * Refueling, maintenance
   * Security

```python
flights_df = pd.read_csv('/home/jewelle/data_bootcamp/flights_df_clean.csv')
```

```python
flights_df.head()
```

```python
#only need flights that were delayed, so remove rows where dep_delay and arr_delay have 0.0 values
flights_df.loc[(flights_df[['dep_delay', 'arr_delay']] != 0).all(axis=1)]
flights_df.head()
```

```python
flights_df.dtypes
```

```python
#convert fl_date to datetime
flights_df['fl_date'] = pd.to_datetime(flights_df['fl_date'], format='%Y-%m-%d')
```

```python
#need to get delay data by month/year, so separate these into new columns
flights_df['fl_date'].dt.year.unique() #This subset only has year 2019, but OG data covers 2018-2019
```

```python
#new columns for year and month
flights_df['year'] = flights_df['fl_date'].dt.year
flights_df['month'] = flights_df['fl_date'].dt.month
```

```python
flights_df.head()
```

```python
#average dep_delays over year
avg_dep_delay = flights_df.groupby('year')['dep_delay'].mean()
med_dep_delay = flights_df.groupby('year')['dep_delay'].median()
#specify year when daataset has more than one
print(avg_dep_delay)
print(med_dep_delay)
#in minutes
```

```python
#average arr_delays over year
avg_arr_delay = flights_df.groupby('year')['arr_delay'].mean()
med_arr_delay = flights_df.groupby('year')['arr_delay'].median()
#specify year when dataset has more than one
print(avg_arr_delay)
print(med_arr_delay)
#in minutes
```

```python
#average dep_delay by month
dep_delay_months = flights_df.groupby('month')['dep_delay'].mean()
#Again, would need to separate by year on full dataset
dep_delay_months
```

```python
#plot dep_delay_months
ax = sns.barplot(x=dep_delay_months.values, 
                 y=[calendar.month_name[i] for i in dep_delay_months.index])

ax.set(title='Average Departure Delay by Month', xlabel='Average Departure Delay in minutes');
#Not super helpful, since my subset is only has data from March 2019.  But the idea is there. 
```

```python
#average dep_delay by month
arr_delay_months = flights_df.groupby('month')['arr_delay'].mean()
#Again, would need to separate by year on full dataset
arr_delay_months
```

```python
#plot dep_delay_months
ax = sns.barplot(x=arr_delay_months.values, 
                 y=[calendar.month_name[i] for i in arr_delay_months.index])

ax.set(title='Average Departure Delay by Month', xlabel='Average Departure Delay in minutes');
#Not super helpful again, since my subset is only has data from March 2019.  But the idea is there. 
```
