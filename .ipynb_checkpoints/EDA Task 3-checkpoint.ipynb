```python
path = '/home/jewelle/data_bootcamp/'
flights_df =  pd.read_csv(path +'flights_sub.csv', parse_dates=['fl_date'])
flights_df.head()
```

```python
flights_df.drop(['Unnamed: 0', 'mkt_unique_carrier', 'dup', 'no_name'], axis=1, inplace=True)
```

```python
# Get all the unique cities
cities = flights_df['origin_city_name'].unique()
flights_df['city'] = flights_df['origin_city_name'].apply(lambda x: x.split(', ')[0])
```

```python
flights_df['country'] = 'US'
```

```python
def convert_time(x):
    if np.isnan(x):
        return '00:00'
    
    if x == 2400:
        return '23:59'   
    string = str(int(x))
    
    if len(string)< 4:
        for _ in range(4-len(string)):
            string = '0' + string
    
    return string[:2] + ':' + string[2:]
```

```python
flights_df['dep_time'] = flights_df['dep_time'].apply(lambda x: convert_time(x))
flights_df['dep_time'] = pd.to_datetime(flights_df['dep_time'], format ='%H:%M')
flights_df['start'] = flights_df['dep_time'].dt.hour
flights_df['arr_time'] = flights_df['arr_time'].apply(lambda x: convert_time(x))
flights_df['arr_time'] = pd.to_datetime(flights_df['arr_time'], format ='%H:%M')
flights_df['end'] = flights_df['arr_time'].dt.hour
```

```python
flights_df['timestamp'] = flights_df['fl_date'].replace(tzinfo=timezone.utc).timestamp()
```
