#### **Task 7**: Test the hypothesis whether planes fly faster when there is the departure delay? 

```python
path = '/home/jewelle/data_bootcamp/'
flights_df =  pd.read_csv(path +'flights_df_clean.csv')
flights_df.head()
```

```python
flights_df = flights_df.drop(flights_df[(flights_df['air_time']==0.0)].index)
```

```python
#check there are no airtime values of 0 again
flights_df[(flights_df['air_time']==0.0)]
```

```python
#speed = distance/airtime
#data is in miles per minute
#miles:km is 1:1.609

flights_df['avg_speed'] = (flights_df['distance']/flights_df['air_time'])*1.61*60 #convert minutes to hours
```

```python
flights_df['avg_speed']
```

```python
#where dep_delay > 0
delay_flightspeed = flights_df.loc[(flights_df['dep_delay'] > 0), 'avg_speed']
```

```python
#where dep_delay = 0
delay_zero_speed = flights_df.loc[(flights_df['dep_delay'] == 0), 'avg_speed']
```

```python
#We take the HNull that there is no difference in average(mean) flight speed with regards to dep_delay
def filter_outliers(data, m = 3):
    return data[abs(data - np.mean(data)) < m * np.std(data)]
```

```python
delay_flightspeed_filtered = filter_outliers(delay_flightspeed, m = 3)
delay_zero_speed_filtered = filter_outliers(delay_zero_speed, m = 2)
```

```python
delay_flightspeed_filtered.mean(), delay_zero_speed_filtered.mean()
```

```python
#Z-test
from statsmodels.stats.weightstats import ztest
ztest, pval = ztest(delay_flightspeed_filtered, delay_zero_speed_filtered)
```

```python
print("p-value",pval)
if pval < 0.05:
      print("we reject null hypothesis")
else:
      print("we accept null hypothesis")
```

```python
#Two-sample t-test
from scipy.stats import ttest_ind
ttest, pval = ttest_ind(delay_flightspeed_filtered[:1000], delay_zero_speed_filtered[:1000])
```

```python
print("p-value",pval)
if pval < 0.05:
      print("we reject null hypothesis")
else:
      print("we accept null hypothesis")
```

Reasons why they might not be faster:
* Delays due to inclement weather do not disappear after take-off. 
* Delays due to large payload (large volume of passengers + cargo) aren't going to help flight speed
* Fuel Limitations
