#### **Task 8**: When (which hour) do most 'LONG', 'SHORT', 'MEDIUM' haul flights take off?

```python
path = '/home/jewelle/data_bootcamp/'
flights_df =  pd.read_csv(path +'flights_sub.csv')
flights_df.head()
```

```python
flights_df.drop(['Unnamed: 0', 'dup', 'mkt_unique_carrier', 'no_name'], axis = 1)
```

Short, medium, and long haul flights can be defined by either distance or travel time.  

One definition by distance:
* short: between 700 - 2600 miles/1120 -  4186 km
* medium: 2,600 - 3000 miles/4186 - 4830 km
* long: over 3,000 miles/4830+ km

https://en.wikipedia.org/wiki/Flight_length#North_America

If we look at time instead:
* short: 30 mins - 3 hours
* medium: 3 -6 hours
* long: 6+ hours

https://www.mcginleygroup.co.uk/blog/the-differences-between-longhaul-and-shorthaul-piloting/bp67/#:~:text=Short%2Dhaul%20is%20a%20flight,that%20extend%20beyond%206%20hours.

```python
flights_df = flights_df.loc[(flights_df['air_time'].notna()) & (flights_df['cancelled'] == 0)]
```

```python
#convert to datetime
def convert_time(x):
    if np.isnan(x):
        return '00:00'
    
    if x == 2400:
        return '23:59'   
    string = str(int(x))
    
    if len(string)< 4:
        for _ in range(4-len(string)):
            string = '0' + string
    
    return string[:2] + ':' + string[2:]
```

```python
#get dep_hr again
flights_df['fl_date'] = pd.to_datetime(flights_df['fl_date'], format='%Y-%m-%d')
```

```python
flights_df['dep_time'] = flights_df['dep_time'].apply(lambda x: convert_time(x))
flights_df['dep_time'] = pd.to_datetime(flights_df['dep_time'], format='%H:%M')
flights_df['dep_hr'] = flights_df['dep_time'].dt.hour
```

```python
#Taking distance as a variable to measure type of haul
#check values
flights_df['distance'].isna().sum()
```

```python
flights_df[(flights_df['distance']==0.0)]
```

```python
#miles:km is 1:1.61
def haul_distance(x):
    if x < 2600:
        return 'SHORT'
    elif x >= 2600 and x < 3000:
        return 'MEDIUM'
    elif x >= 3000:
        return 'LONG'
```

```python
flights_df['haul'] = flights_df['distance'].apply(lambda x: haul_distance(x))
```

```python
flights_df.head()
```

```python
x = flights_df.groupby(['dep_hr', 'haul']).count()
```

```python
x
```

```python
x.reset_index(inplace=True)
```

```python
sns.set_style("whitegrid")

ax1 = sns.barplot(x ='dep_hr', y='fl_date', data = x.loc[x['haul'] == 'SHORT'], color = 'red');

ax1.set(title='Short Haul Flights Departing By Hour (24hr)', xlabel='Hour(24hrs)', ylabel='Number of Flights')
```

```python
ax1 = sns.barplot(x ='dep_hr', y='fl_date', data = x.loc[x['haul'] == 'MEDIUM'], color = 'blue');

ax1.set(title='Medium Haul Flights Departing By Hour(24hr)', xlabel='Hour(24hrs)', ylabel='Number of Flights')
```

```python
ax1 = sns.barplot(x ='dep_hr', y='fl_date', data = x.loc[x['haul'] == 'LONG'], color = 'green');

ax1.set(title='Long Haul Flights Departing By Hour (24hr)', xlabel='Hour(24hrs)', ylabel='Number of Flights')
```

```python
#Taking time as a variable to measure haul
def haul_time(x):
    x = x/60
    if x < 3:
        return 'SHORT'
    elif x >= 3 and x < 6:
        return 'MEDIUM'
    elif x >= 6:
        return 'LONG'
```

```python
flights_df['haul_time'] = flights_df['air_time'].apply(lambda x: haul_time(x))
```

```python
x = flights_df.groupby(['dep_hr', 'haul_time']).count()
x
```

```python
x.reset_index(inplace=True)
```

```python
ax1 = sns.barplot(x ='dep_hr', y='fl_date', data = x.loc[x['haul_time'] == 'SHORT'], color = 'red');

ax1.set(title='Short Haul Flights Departing By Hour (24hr)', xlabel='Hour(24hrs)', ylabel='Number of Flights')
```

```python
ax1 = sns.barplot(x ='dep_hr', y='fl_date', data = x.loc[x['haul_time'] == 'MEDIUM'], color = 'blue');

ax1.set(title='Medium Haul Flights Departing By Hour(24hr)', xlabel='Hour(24hrs)', ylabel='Number of Flights')
```

```python
ax1 = sns.barplot(x ='dep_hr', y='fl_date', data = x.loc[x['haul_time'] == 'LONG'], color = 'green');

ax1.set(title='Long Haul Flights Departing By Hour (24hr)', xlabel='Hour(24hrs)', ylabel='Number of Flights')
```
