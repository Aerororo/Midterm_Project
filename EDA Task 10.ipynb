#### **Task 10**: Do bigger delays lead to bigger fuel comsumption per passenger? 
We need to do four things to answer this as accurate as possible:
- Find out average monthly delay per air carrier (monthly delay is sum of all delays in 1 month)

```python
path = '/home/jewelle/data_bootcamp/'
flights_df =  pd.read_csv(path +'flights_df_clean.csv', parse_dates=['fl_date'])
flights_df.head()
```

```python
#flights_df['fl_date'] = pd.to_datetime(flights_df['fl_date'], infer_datetime_format=True)
#unique carrier id shared by all three tables
```

```python
flights_df['month'] = flights_df['fl_date'].dt.month
```

```python
#delays by month per unique air carrier
delay_month = pd.DataFrame(flights_df.loc[(flights_df['dep_delay'].notna())].groupby(['month','mkt_carrier'])
                           ['dep_delay'].mean().reset_index())
```

```python
delay_month.head()
```

- Find out distance covered monthly by different air carriers

```python
#distance by month per unique air carrier
distance_month = pd.DataFrame(flights_df.groupby(['month', 'mkt_carrier'])['distance'].sum().reset_index())
```

```python
distance_month.head()
```

- Find out number of passengers that were carried by different air carriers

```python
passengers_df = pd.read_csv(path +'passengers_sub.csv')
passengers_df.head()
```

```python
passengers_df.drop(['Unnamed: 0'], axis=1)
```

```python
#passengers per unique carrier
p_carrier = pd.DataFrame(passengers_df.loc[passengers_df
                                           ['unique_carrier'].isin
                                           (distance_month['mkt_carrier'].unique())].groupby('unique_carrier')
                         ['passengers'].sum().reset_index())
```

```python
p_carrier.head()
```

- Find out total fuel comsumption per air carrier.

```python
fuel_df = pd.read_csv(path +'fuel_consumption_sub.csv')
fuel_df.head()
```

```python
fuel_df.drop(['Unnamed: 0'], axis=1)
```

```python
#fuel consumption in gallons per unique carrier
con_carrier = pd.DataFrame(fuel_df.loc[fuel_df
                                       ['unique_carrier'].isin(p_carrier
                                                               ['unique_carrier'].values)].groupby
                           ('unique_carrier')['tdomt_gallons'].sum().reset_index())
```

Use this information to get the average fuel comsumption per passenger per km. Is this higher for the airlines with bigger average delays?

```python
x = pd.DataFrame(delay_month.groupby('mkt_carrier')
                 ['dep_delay'].sum().reset_index()).rename({'dep_delay': 'dep_delay_monthly_avg', 
                                                            'mkt_carrier': 'unique_carrier'}, axis=1)
```

```python
y = pd.DataFrame(distance_month.groupby('mkt_carrier')
                 ['distance'].sum().reset_index()).rename({'distance': 'distance_monthly_avg', 
                                                           'mkt_carrier': 'unique_carrier'}, axis=1)
```

```python
from functools import reduce
```

```python
final = reduce(lambda x,y: pd.merge(x,y, on='unique_carrier', how='outer'), [x, y, con_carrier, p_carrier])
```

```python
final.head()
```

```python
final['fuel_com_final'] = final.apply(lambda row: 
                                      (row['tdomt_gallons']/row['passengers'])/row['dep_delay_monthly_avg'], axis=1)
```

```python
final.sort_values('fuel_com_final', ascending=False)
```

```python
axis = plt.subplots(figsize=(14,10))
ax = sns.lineplot(x = 'dep_delay_monthly_avg', y = 'fuel_com_final', data = final)
ax.set(xlabel='Average Departure Delay Per Month ', ylabel='Fuel Consumption Per Passenger Per Mile')
```
