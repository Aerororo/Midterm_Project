Short, medium, and long haul flights can be defined by either distance or travel time.  

One definition of distance:
* short: between 700-2600 miles
* medium: 2,600 - 3000 miles
* long: over 3,000 mi

https://en.wikipedia.org/wiki/Flight_length#North_America

If we look at time instead:
* short: 30 mins - 3 hours
* medium: 3 -6 hours
* long: 6+ hours
